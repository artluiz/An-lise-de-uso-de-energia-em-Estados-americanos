```python
import numpy as np
import pandas as pd
import pandasql as ps

import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


import tensorflow as tf
##import tensorflow_addons as tfa
from tensorflow_addons.metrics import RSquare

from google.colab import files
import io
```

```python
Sales = pd.read_csv(io.BytesIO(uploaded['vgsales.csv']), index_col='Rank')

Sales
```

```python
drop_col = ['Name', 'NA_Sales',	'EU_Sales',	'JP_Sales',	'Other_Sales']
Sales.drop(drop_col, axis=1, inplace=True)

Sales
```

```python
Sales.isnull().sum()
```

```python
Sales['Year'] = Sales['Year'].fillna(Sales['Year'].mean())
```

```python
Sales = Sales.dropna(axis=0)
```

```python
Sales['Platform'].unique()
```

```python
Sales['Genre'].unique()
```

```python
counts = Sales['Publisher'].value_counts()
Sales['Publisher'] = Sales['Publisher'].apply(lambda x: 'Small Publisher' if counts[x] < 50 else x)
```

```python
Sales['Publisher'].value_counts()
```

```python
onehot_col = ['Platform', 'Genre', 'Publisher']
```

```python
def oh_code(Sales, columns):
    for column in columns:
        dummies = pd.get_dummies(Sales[column])
        Sales = pd.concat([Sales, dummies], axis = 1)
        Sales.drop(column, axis = 1, inplace=True)
    return Sales
```

```python
Sales = oh_code(Sales, onehot_col)
Sales
```

```python
y = Sales['Global_Sales']
x = Sales.drop('Global_Sales', axis = 1)

x
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)

x.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8)
```

```python
from keras.optimizers.legacy.rmsprop import RMSProp
input = tf.keras.Input(shape = (91, ))
x = tf.keras.layers.Dense(128, activation = 'relu')(input)
x = tf.keras.layers.Dense(128, activation = 'relu')(x)
output = tf.keras.layers.Dense(1)(x)

model = tf.keras.Model(inputs = input, outputs = output)

optimizer = tf.keras.optimizers.RMSprop(0.001)

model.compile(
    optimizer=optimizer,
    loss='mse'
)

batch_size = 64
epochs = 300

history = model.fit(
    x_train,
    y_train,
    validation_split = 0.2,
    batch_size = batch_size,
    epochs = epochs,
    verbose = 0
)
```

```python
plt.figure(figsize=(14, 10))

epochs_range = range(1, epochs +1)
train_loss = history.history['loss']
val_loss = history.history['val_loss']

plt.plot(epochs_range, train_loss, label = 'Training loss')
plt.plot(epochs_range, val_loss, label = 'Validation loss')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
y_pred = np.squeeze(model.predict(x_test))

result = RSquare()
result.update_state(y_test, y_pred)

print('R^2 Score:', result.result())
```
